```python
# –¢–µ–ø–ª–æ–≤–∞ –∫–∞—Ä—Ç–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—ó —á–∏—Å–ª–æ–≤–∏—Ö –æ–∑–Ω–∞–∫
corr = df.select_dtypes(include=['float64', 'int64']).corr().round(2)
fig = px.imshow(corr,
                text_auto=True,
                title='–¢–µ–ø–ª–æ–≤–∞ –∫–∞—Ä—Ç–∞ –∫–æ—Ä–µ–ª—è—Ü—ñ—ó',
                color_continuous_scale='blues')
fig.show()
```

–û–±—á–∏—Å–ª–∏–º–æ –º–µ—Ç—Ä–∏–∫–∏ –æ—Ü—ñ–Ω–∫–∏ —è–∫–æ—Å—Ç—ñ –º–æ–¥–µ–ª—ñ:

- **MAE (Mean Absolute Error)** ‚Äî —Å–µ—Ä–µ–¥–Ω—è –∞–±—Å–æ–ª—é—Ç–Ω–∞ –ø–æ—Ö–∏–±–∫–∞.
- **RMSE (Root Mean Squared Error)** ‚Äî –∫–æ—Ä—ñ–Ω—å —Å–µ—Ä–µ–¥–Ω—å–æ—ó –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ—ó –ø–æ—Ö–∏–±–∫–∏.
- **R¬≤ (R-squared)** ‚Äî –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –¥–µ—Ç–µ—Ä–º—ñ–Ω–∞—Ü—ñ—ó (–ø–æ—è—Å–Ω–µ–Ω–æ—ó –¥–∏—Å–ø–µ—Ä—Å—ñ—ó).

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2
```

```python
train_mae, train_rmse, train_r2 = regression_metrics(y_train, y_train_pred)
test_mae, test_rmse, test_r2 = regression_metrics(y_test, y_test_pred)

print(f"Train MAE: {train_mae:.4f}, RMSE: {train_rmse:.4f}, R2: {train_r2:.4f}")
print(f"Test MAE: {test_mae:.4f}, RMSE: {test_rmse:.4f}, R2: {test_r2:.4f}")
```
