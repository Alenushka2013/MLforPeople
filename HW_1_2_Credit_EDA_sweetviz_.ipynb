# Додаткові бібліотеки для EDA аналізу

**Sweetviz**:

Генерує красиві, інтерактивні HTML-звіти з візуалізаціями для кожного стовпця даних.

Дозволяє легко порівнювати два набори даних (наприклад, навчальні та тестові).

Фокусується на виявленні взаємозв'язків між змінними.

Дуже простий у використанні, всього кілька рядків коду.

**Pandas Profiling:**

Ще одна дуже популярна бібліотека для автоматичного створення детальних інтерактивних HTML-звітів.

Надає широкий спектр статистики: розподіли, кореляції, пропущені значення, унікальні значення тощо.

Добре інтегрується з pandas DataFrame.

Надає більше деталей про різні типи кореляцій (Пірсона, Спірмена, Кендалла).

**D-Tale:**

Надає інтерактивний веб-інтерфейс для перегляду та аналізу DataFrame прямо у Jupyter Notebook або як окремий застосунок.

Дозволяє сортувати, фільтрувати, будувати графіки, переглядати статистику, виявляти пропущені значення та багато іншого, не пишучи багато коду. Це дуже зручно для візуального дослідження даних.

**DataPrep**:

Пропонує низку функцій для підготовки даних, включаючи create_report() для генерації інтерактивних звітів.

Може бути швидшим для великих наборів даних завдяки оптимізації.

**AutoViz:**

Як випливає з назви, AutoViz фокусується на автоматичній візуалізації даних.

Може допомогти швидко виявити найбільш впливові ознаки та побудувати різноманітні графіки.

*Яку вибрати?*

Якщо вам потрібен швидкий, візуально привабливий звіт з акцентом на порівняння наборів даних та взаємозв'язки, Sweetviz є чудовим вибором.

Якщо ви хочете дуже детальний звіт з усіма статистичними даними, кореляціями та виявленням проблем з даними, Pandas Profiling часто є кращим варіантом.

Якщо ви віддаєте перевагу інтерактивному графічному інтерфейсу для дослідження даних, спробуйте D-Tale.

Для швидкої автоматичної візуалізації може бути корисним AutoViz.
